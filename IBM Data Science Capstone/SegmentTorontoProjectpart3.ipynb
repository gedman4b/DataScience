### Retrieve data from Wiki page

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
response = requests.get(url)
response.status_code
```

### Scrape Web page content

```python
soup = BeautifulSoup(response.text, "html.parser")
table = soup.find('table', attrs={'class': 'wikitable sortable'})
table_rows = table.find_all('tr')

res = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        res.append(row)
```

### Tranform the data into a pandas dataframe

```python
# define the dataframe columns
column_names = ['PostalCode', 'Borough', 'Neighborhood'] 

# instantiate the dataframe
df = pd.DataFrame(res, columns=column_names)
df.head(10)
```

```python
# get row count for Not Assigned Boroughs
df[df['Borough'] != 'Not assigned'].count()
```

### Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
df = df[df['Borough'] != 'Not assigned']
df.head(10)
```

### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. So for the 9th cell in the table on the Wikipedia page, the value of the Borough and the Neighborhood columns will be Queen's Park.

```python
df = df.mask(df == 'Not assigned', df['Borough'], axis=0)
df.head(20)
```

### If a zip code is listed twice these two rows will be combined into one row with the neighborhoods separated with a comma

```python
df = df.groupby(['PostalCode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
df
```

### Import Geospatial data

```python
dfll = pd.read_csv('Geospatial_Coordinates.csv')
dfll
```

### Merge the Toronto Neighborhood dataframe with the Geospaitial dataframe

```python
dfll.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
df_segments = pd.merge(df, dfll, on='PostalCode', how='inner')
df_segments
```

### Use geopy library to get the latitude and longitude values of Toronto.

```python
address = 'Toronto, CA'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

### Create map of Toronto using latitude and longitude values

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_segments['Latitude'], df_segments['Longitude'], df_segments['Borough'], df_segments['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
